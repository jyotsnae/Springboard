```python
df1 = pd.get_dummies(df)

df1.head(10)
```

```python
X = df1
#removing the target column revenue from x
X = X.drop(['Revenue'],axis = 1)

#fixing target variable
y = df['Revenue']
```

```python
# splitting the data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,roc_curve,roc_auc_score
from sklearn.metrics import accuracy_score
from matplotlib import pyplot
lr = LogisticRegression()
model_lr = lr.fit(X_train,y_train)
y_pred_lr = model_lr.predict(X_test)

#Evaluating model
print('\n Confusion Matrix \n',confusion_matrix(y_pred_lr,y_test))
print("\n Accuracy of Logistic Regression: ",accuracy_score(y_test,y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred_rf = model.predict(X_test)

print('\n Confusion Matrix \n',confusion_matrix(y_pred_rf,y_test))
print("\n Accuracy of Random Forest: ",accuracy_score(y_test,y_pred_rf))
```

```python
from sklearn.svm import SVC
svc = SVC()
model = svc.fit(X_train,y_train)
y_pred_svc = model.predict(X_test)


print('\n Confusion Matrix \n',confusion_matrix(y_pred_svc,y_test))
print("\n Accuracy of SVM: ",accuracy_score(y_test,y_pred_svc))
```
