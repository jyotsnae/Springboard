# NUTRITION FACTS OF STARBUCKS DRINKS

This dataset includes the nutritional information for Starbucks drink menu items. It contains 18 columns and 242 rows. For this data analysis, I am only including the Grande size drinks.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("starbucks_drinkMenu_expanded.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.columns = ['Category','Beverage','Bev_prep','Calories','Total_fat','Trans_fat','Sat_fat','Sodium','Total_carbs','Cholesterol','Fibre','Sugars','Protein','VitA','VitC','Calcium','Iron','Caffeine']
```

```python
data = data.dropna()
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
# How many different beverages are there in each category?
data.groupby("Category")["Beverage"].nunique().plot(kind='bar', figsize=(8, 6))
plt.title('Variety in Starbucks Drinks', fontsize=16)
plt.xlabel("Drink Category")
plt.ylabel("Number of Beverages")
plt.show()
```

```python
# Top 10 drinks with the highest calories
data.nlargest(10, 'Calories')[['Beverage', 'Calories']]
```

```python
#Distribution of Calories
plt.hist(data.Calories, bins = 6, rwidth= 0.85)
plt.title('Distribution of Calories')
plt.xlabel("Calories")
plt.ylabel("Number of Drinks")
plt.show()
```

```python
# Top 10 drinks with the most sugar
data.nlargest(10, 'Sugars')[['Beverage', 'Sugars']]
```

```python
#Drinks with the most Protein
mostprotein=data[data['Protein']>=18]
mostprotein
```

```python
#Drinks with less than 200 calories and less than 20mgms of sugar
caloriesless160=data[(data['Calories'] < 200) & (data['Sugars'] < 20)]
caloriesless160.head()
```

```python
#correlation of nutritional attributes
plt.figure(figsize=(10,8), dpi= 80)
sns.heatmap(data.corr(), xticklabels=data.corr().columns, yticklabels=data.corr().columns, annot = True)
plt.title('Correlation of Nutritional Attributes', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
